### 4a

```python
X_train_4 = train_df[["Sleep_Duration"]].values
Y_train4 = train_df["Sleep_Quality"].values.reshape(-1, 1)

p = 3

X_poly_train4 = np.hstack([X_train_4**k for k in range(1, p+1)])

ones = np.ones((X_poly_train4.shape[0], 1))
X_poly_train_b4 = np.hstack([ones, X_poly_train4])

theta_poly = np.linalg.pinv(X_poly_train_b4.T @ X_poly_train_b4) @ X_poly_train_b4.T @ Y_train4

y_train_pred4 = X_poly_train_b4 @ theta_poly

train_mse4 = mean_squared_error(Y_train4, y_train_pred4)
train_r4 = r2_score(Y_train4, y_train_pred4)

print("Polynomial Regression")
print(f"Degree p = {p}")
print(f"Train MSE: {train_mse4} | Train R2: {train_r4}")
```

I implemented polynomial regression by taking the single feature Sleep Duration and expanding it into multiple features using powers of that variable. I then added a column of ones to include the intercept term. After constructing this new feature matrix, I applied the same closed-form least-squares solution from Problem 3 to compute the coefficients. Finally, I generated predictions using matrix multiplication and evaluated the model using MSE and R2.

### 4b

```python
X_test_4 = test_df[["Sleep_Duration"]].values
Y_test4 = test_df["Sleep_Quality"].values.reshape(-1, 1)

results = []

for p in [1, 3, 5]:
    X_poly_train4 = np.hstack([X_train_4**k for k in range(1, p+1)])
    X_poly_test4  = np.hstack([X_test_4**k  for k in range(1, p+1)])

    X_poly_train_b4 = np.hstack([np.ones((X_poly_train4.shape[0], 1)), X_poly_train4])
    X_poly_test_b4  = np.hstack([np.ones((X_poly_test4.shape[0], 1)),  X_poly_test4])

    theta_poly = np.linalg.pinv(X_poly_train_b4.T @ X_poly_train_b4) @ X_poly_train_b4.T @ Y_train4

    y_train_pred4 = X_poly_train_b4 @ theta_poly
    y_test_pred4  = X_poly_test_b4 @ theta_poly

    train_mse4 = mean_squared_error(Y_train4, y_train_pred4)
    train_r4   = r2_score(Y_train4, y_train_pred4)

    test_mse4  = mean_squared_error(Y_test4, y_test_pred4)
    test_r4    = r2_score(Y_test4, y_test_pred4)

    results.append([p, train_mse4, train_r4, test_mse4, test_r4])

print("p | Train MSE | Train R2 | Test MSE | Test R2")
print("--|----------|----------|----------|---------")
for p, tr_mse, tr_r2, te_mse, te_r2 in results:
    print(f"{p}  {round(tr_mse,7)} | {round(tr_r2,7)} | {round(te_mse,7)} | {round(te_r2,7)}")
```

As the polynomial degree p increases from 1 to 5, the training MSE decreases slightly and the training R2 increases, which shows that the model fits the training data better with higher-degree terms. The test MSE also decreases and the test R2 increases across these values of p, meaning the model is also improving on unseen data. This suggests that the relationship between sleep duration and sleep quality is not perfectly linear and that adding polynomial terms up to degree 5 helps capture that pattern without overfitting.

## 5a

```python
def addIntercept(X):
    return np.hstack([np.ones((X.shape[0], 1)), X])

def r2Score(Y_true, Y_pred):
    ss_res = np.sum((Y_true - Y_pred) ** 2)
    ss_tot = np.sum((Y_true - np.mean(Y_true)) ** 2)
    return 1 - ss_res / ss_tot

def normalizeTrainTest(X_train, X_test):
    mean = np.mean(X_train, axis=0)
    std = np.std(X_train, axis=0)
    std[std == 0] = 1

    X_train_norm = (X_train - mean) / std
    X_test_norm = (X_test - mean) / std

    return X_train_norm, X_test_norm

def gradientDescent(X_train, Y_train, alpha, num_iters):
    m = X_train.shape[0]
    theta = np.zeros((X_train.shape[1], 1))
    mse_history = []

    for i in range(num_iters):
        Y_pred = X_train @ theta
        gradient = (1/m) * (X_train.T @ (Y_pred - Y_train))
        theta = theta - alpha * gradient
        mse_history.append(mean_squared_error(Y_train, Y_pred))

    return theta, mse_history
```

```python
Xtrain5 = train_df.drop(columns=["Sleep_Quality"]).values
Ytrain5 = train_df["Sleep_Quality"].values.reshape(-1, 1)

Xtest5 = test_df.drop(columns=["Sleep_Quality"]).values
Ytest5 = test_df["Sleep_Quality"].values.reshape(-1, 1)

XtrainB5 = addIntercept(Xtrain5)

thetaNoNorm5, mseNoNorm5 = gradientDescent(
    XtrainB5, Ytrain5, alpha=0.000001, num_iters=100
)

YtrainPred5 = XtrainB5 @ thetaNoNorm5
print("No normalization")
print("Train MSE:", mean_squared_error(Ytrain5, YtrainPred5))
print("Train R2:", r2Score(Ytrain5, YtrainPred5))

XtrainNorm5, XtestNorm5 = normalizeTrainTest(Xtrain5, Xtest5)

XtrainNormB5 = addIntercept(XtrainNorm5)
XtestNormB5 = addIntercept(XtestNorm5)

thetaNorm5, mseNorm5 = gradientDescent(
    XtrainNormB5, Ytrain5, alpha=0.1, num_iters=100
)

YtrainPredNorm5 = XtrainNormB5 @ thetaNorm5
print("\nWith normalization")
print("Train MSE:", mean_squared_error(Ytrain5, YtrainPredNorm5))
print("Train R2:", r2Score(Ytrain5, YtrainPredNorm5))
```

I implemented gradient descent by initializing the parameter vector to zeros and iteratively updating it using the gradient of the mean squared error loss. I first applied the algorithm without normalizing the features, which caused an error. I then normalized each feature by centering it at the mean and scaling by the standard deviation, which stabilized the updates and allowed the algorithm to converge to a good solution.

## 5b

```python
alphas = [0.01, 0.1, 0.5]
iters = [10, 50, 100]

print("alpha iters | Train MSE | Train R2 | Test MSE | Test R2")
print("------------|----------|----------|----------|---------")

for alpha in alphas:
    for i in iters:
        theta5, _ = gradientDescent(XtrainNormB5, Ytrain5, alpha, i)

        YtrainPred5 = XtrainNormB5 @ theta5
        YtestPred5 = XtestNormB5 @ theta5

        trainMse5 = mean_squared_error(Ytrain5, YtrainPred5)
        trainR25 = r2Score(Ytrain5, YtrainPred5)

        testMse5 = mean_squared_error(Ytest5, YtestPred5)
        testR25 = r2Score(Ytest5, YtestPred5)

        print(
            f"{alpha} {i:>5} | "
            f"{round(trainMse5,7)} | "
            f"{round(trainR25,7)} | "
            f"{round(testMse5,7)} | "
            f"{round(testR25,7)}"
        )
```

As the number of iterations increases, the training and testing MSE generally decrease and the R2 increases. Smaller learning rates converge more slowly, while larger learning rates converge faster but can be less stable.

## 5c

The learning rate has a big impact on how fast gradient descent converges. With a small learning rate (α = 0.01), the algorithm converges very slowly, and even after 100 iterations the MSE is still high and the R² values are negative. This shows the model is not fitting the data well yet. With a moderate learning rate (α = 0.1), the error drops much faster. By 50 iterations, the MSE is low and the R2 is close to 0.9 on both the training and testing sets. A larger learning rate (α = 0.5) converges even faster. It achieved similar performance in fewer iterations, but shows small changes as the number of iterations increases. Overall, once the features are normalized and the learning rate is reasonable, gradient descent converges to a solution that is very close to the closed-form model.
