```python
import numpy as np

for i in range(3):
    A = np.random.randint(-10, 11, size=(3, 3))
    print("Matrix:")
    print(A)
    
    det = np.linalg.det(A)
    
    if det != 0:
        A_inv = np.linalg.inv(A)
        print("Inverse:")
        print(A_inv)
    
        print("A * A_inv:")
        print(A @ A_inv)
    else:
        print("This matrix is not invertible.")
```

```python
import pandas as pd

# load the dataset
df = pd.read_csv("kc_house_data.csv")

# drop ignored columns
df = df.drop(columns=["id", "date", "zipcode"])

# (a)

stats = []

for col in df.columns:
    stats.append({
        "feature": col,
        "mean": df[col].mean(),
        "min": df[col].min(),
        "max": df[col].max(),
        "variance": df[col].var()
    })

stats_df = pd.DataFrame(stats)

print("(a): Mean, Min, Max, Variance\n", stats_df)

# get features with the lowest and highest mean
lowest_mean = stats_df.loc[stats_df["mean"].idxmin()]
highest_mean = stats_df.loc[stats_df["mean"].idxmax()]

print("\nLowest mean feature:")
print(f"{lowest_mean['feature']} | mean = {lowest_mean['mean']}")

print("\nHighest mean feature:")
print(f"{highest_mean['feature']} | mean = {highest_mean['mean']}")

# get features with lowest and highest variance
lowest_var = stats_df.loc[stats_df["variance"].idxmin()]
highest_var = stats_df.loc[stats_df["variance"].idxmax()]

print("\nLowest variance feature:")
print(f"{lowest_var['feature']} | variance = {lowest_var['variance']}")

print("\nHighest variance feature:")
print(f"{highest_var['feature']} | variance = {highest_var['variance']}")


# (b)

corr_df = (
    df.corr(numeric_only=True)["price"]
    .drop("price")
    .reset_index()
)

corr_df.columns = ["feature", "correlation"]
corr_df = corr_df.sort_values(by="correlation", ascending=False)

print("\n(b): Correlation with price")
print(corr_df)

positive_corr = corr_df[corr_df["correlation"] > 0]

print("\nPositively correlated features:")
print(positive_corr)

top_feature = corr_df.iloc[0]
print("\nHighest positive correlation:")
print(f"{top_feature['feature']} | correlation = {top_feature['correlation']}")


# (c)

negative_corr = corr_df[corr_df["correlation"] < 0]

print("\n(c): Negatively correlated features")
if negative_corr.empty:
    print("No features have a negative correlation with price.")
else:
    print(negative_corr)
```
